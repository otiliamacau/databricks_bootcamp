# Reading the data

```python
df = spark.read.option("header", "true").option("inferSchema", "true").csv("/Volumes/workspace/bronze/source_system/crm_source/cust_info.csv", header = True)
df.write.mode("overwrite").saveAsTable("workspace.bronze.cust_info")
```

## Define Ingestion Configuration

```python
INGESTION_CONFIG = [
{
    "source":"crm",
    "path" : "/Volumes/workspace/bronze/source_system/crm_source/cust_info.csv",
    "table" : "crm_cust_info"
},
{
    "source":"crm",
    "path" : "/Volumes/workspace/bronze/source_system/crm_source/prd_info.csv",
    "table" : "crm_prd_info"
},
{
    "source":"crm",
    "path" : "/Volumes/workspace/bronze/source_system/crm_source/sales_details.csv",
    "table" : "crm_sales_details"
},
{
    "source":"erp",
    "path" : "/Volumes/workspace/bronze/source_system/erp_source/CUST_AZ12.csv",
    "table" : "erp_cust_az12"
},
{
    "source":"erp",
    "path" : "/Volumes/workspace/bronze/source_system/erp_source/LOC_A101.csv",
    "table" : "erp_loc_a101"
},
{
    "source":"erp",
    "path" : "/Volumes/workspace/bronze/source_system/erp_source/PX_CAT_G1V2.csv",
    "table" : "erp_px_cat_g1v2"
},
]
```

## Ingest Files into Bronze Tables

```python
for item in INGESTION_CONFIG:
    df = spark.read.option("header", "true").option("inferSchema", "true").csv(item["path"], header = True)
    df.write.mode("overwrite").saveAsTable(f"workspace.bronze.{item["table"]}")
```
