```python
query = """
    SELECT
        s.order_number,
        pr.product_number,
        c.customer_number,
        s.order_date,
        s.ship_date,
        s.due_date,
        s.sales_amount,
        s.quantity,
        s.price
    FROM silver.crm_sales s
    LEFT JOIN gold.dim_products pr
        ON s.product_number = pr.product_key
    LEFT JOIN gold.dim_customers c
        ON s.customer_id = c.customer_id;
"""
df = spark.sql(query)
```

# Wiriting Gold table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
